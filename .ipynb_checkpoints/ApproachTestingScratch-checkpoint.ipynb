##  Experimenting with OpenCV using stock beetle images
First, I'm going to go through the process with a tribolium image I found online

```python
beetles = cv2.imread('Dead_Tribolium.jpg')
copy = beetles.copy()

gray = cv2.cvtColor(beetles, cv2.COLOR_BGR2GRAY)#Convert to grayscale
gray = cv2.bitwise_not(gray)
gray.shape #An image, 682x1024 pixels

plt.imshow(gray, cmap="Greys")

```

```python
thresh = cv2.threshold(gray,0,255,cv2.THRESH_OTSU + cv2.THRESH_BINARY)[1]

plt.imshow(thresh, cmap="Greys")
```

```python
contours = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)[-2]
for contour in contours:
   cv2.drawContours(gray, contour, -1, (0, 255, 0), 3)
plt.figure()
plt.imshow(gray, cmap="Greys")
```

```python
contours = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_TC89_L1 )[-2]
for contour in contours:
   cv2.drawContours(gray, contour, -1, (0, 255, 0), 3)
plt.figure()
plt.imshow(gray, cmap="Greys")
```

```python
contours = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_TC89_KCOS)[-2]
for contour in contours:
   cv2.drawContours(gray, contour, -1, (0, 255, 0), 3)
plt.figure()
plt.imshow(gray, cmap="Greys")
```

```python
beetle1 = cv2.imread('january10/IMG_2278.JPG')

gray = cv2.cvtColor(beetle1, cv2.COLOR_BGR2GRAY)#Convert to grayscale
gray = cv2.bitwise_not(gray)
gray.shape #An image, 682x1024 pixels

plt.imshow(gray, cmap="Greys")
```

We should probably start by masking around the edge of the plat. We can use HoughCircles to do that, allowing us to detect a circle around the plate's circumference that we can use for masking. (This code from https://www.geeksforgeeks.org/circle-detection-using-opencv-python/)

```python
#Started by setting the min radius to be =1 and max to be =30,000, but that took like 3min to run
#Trying again with some more restricted params
detected_circles = cv2.HoughCircles(gray,  
                   cv2.HOUGH_GRADIENT, 1, 20, param1 = 50,#IDK what this is doing 
               param2 = 30, #I also don't know what this is doing
                                    minRadius = 500, maxRadius = 2000) 
```

```python
# Draw circles that are detected. 
if detected_circles is not None: 
  
    # Convert the circle parameters a, b and r to integers. 
    detected_circles = np.uint16(np.around(detected_circles)) 
  

    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
  
        # Draw the circumference of the circle. 
        cv2.circle(beetle1, (a, b), r, (0, 255, 0), 2) 
  
        # Draw a small circle (of radius 1) to show the center. 
        cv2.circle(beetle1, (a, b), 1, (0, 0, 255), 3) 
        cv2.imshow("Detected Circle", beetle1) 
        cv2.waitKey(0) 
```
